````python
from langchain_community.document_loaders import WebBaseLoader
````

````python
loader = WebBaseLoader('https://careers.nike.com/entry-data-engineer/job/R-45234?source=BY_LinkedIn_OrganicFeed&utm_source=BY_LinkedIn_OrganicFeed')
docs = loader.load().pop().page_content
print(docs)
````

````python
from langchain_core.prompts import PromptTemplate
````

````python

prompt_extract = PromptTemplate.from_template(
       """
    ### SCRAPED TEXT FROM WEBSITE:
    {page_data}
    
    ### INSTRUCTION:
    Extract job postings and return them in **valid JSON format** with the following keys:
    - `role`: The job title.
    - `experience`: Years of experience required as an integer.
    - `skills`: A list of required skills.
    - `description`: A **detailed** job description (at least 3 sentences).

    **Ensure the JSON is properly formatted and provide a complete description.**
    Only return valid JSON without extra text.
    """)

chain_extract = prompt_extract | chat 
res = chain_extract.invoke(input={'page_data':docs})
print(res.content)
````

````python
import json

def clean_json_response(response):
    # Remove markdown code block markers if present
    if response.startswith("```json"):
        response = response[7:].strip()  # Remove "```json" and strip whitespace
    if response.endswith("```"):
        response = response[:-3].strip()  # Remove closing triple backticks and strip whitespace
    
    return response

# Example LLM response with JSON formatting
response_text = res.content
# Clean the response and parse it as a dictionary
cleaned_response = clean_json_response(response_text)
parsed_dict = json.loads(cleaned_response)  # Convert to Python dictionary

# Print dictionary (not as JSON)
print(parsed_dict)
````
